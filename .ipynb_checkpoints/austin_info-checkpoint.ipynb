```python
#importing dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
austin_data = "aac_intakes.csv"
austin_df = pd.read_csv(austin_data)
austin_df.head()
```

```python
# Create a filtered dataframe from specific columns
austin_cols = ['animal_id', 'animal_type', 'breed', 'datetime', 'found_location', 'intake_condition', 'intake_type']
austin_transformed= austin_df[austin_cols].copy()

# Rename the column headers
austin_transformed = austin_transformed.rename(columns={"animal_id": "animal_id",
                                                        "animal_type": "animal_type",
                                                        "breed": "animal_breed", 
                                                        "datetime": "intake_date",
                                                        "found_location": "intake_location",
                                                        "intake_condition": "intake_condition", 
                                                        "intake_type": "intake_type"
                                                       })

# Clean the data by dropping duplicates and setting the index
austin_transformed.drop_duplicates("animal_id", inplace=True)
austin_transformed.set_index("animal_id", inplace=True)

austin_transformed.head()
```

```python
#create a database connection
connection_string = "postgres:postgres@localhost:5432/ETL_project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#confirm tables in database
engine.table_names()
```

```python
#push data to Postgres
austin_transformed.to_sql(name='austin_shelter', con=engine, if_exists='append', index=True)
```
